```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('avocado.csv')

print(data.head())
print(data.describe())

#membuat scatter plot
plt.scatter(data['4046'], data['year'])
plt.xlabel('4046')
plt.ylabel('year')
plt.title('Penjualan alpukat seri 4046')
plt.show()
plt.scatter(data['4225'], data['year'])
plt.xlabel('4225')
plt.ylabel('year')
plt.title('Penjualan alpukat seri 4225')
plt.show()
plt.scatter(data['4770'], data['year'])
plt.xlabel('4770')
plt.ylabel('year')
plt.title('Penjualan alpukat seri 4770')
plt.show()

#membuat bar plot
```

```python
import pandas as pd
import seaborn as sns

data = pd.read_csv('avocado.csv')

sns.scatterplot(x='4046', y='year', data=data)
plt.xlabel('4046')
plt.ylabel('year')
plt.title('Penjualan alpukat seri 4046')
plt.show()

sns.barplot(x='type', y='4046', data=data)
plt.xlabel('type')
plt.ylabel('4046')
plt.title('Tipe Paling Laku')
plt.show()

sns.scatterplot(x='4225', y='year', data=data)
plt.xlabel('4225')
plt.ylabel('year')
plt.title('Penjualan alpukat seri 4225')
plt.show()

sns.barplot(x='type', y='4225', data=data)
plt.xlabel('type')
plt.ylabel('4225')
plt.title('Tipe Paling Laku')
plt.show()

sns.scatterplot(x='4770', y='year', data=data)
plt.xlabel('4770')
plt.ylabel('year')
plt.title('Penjualan alpukat seri 4770')
plt.show()

sns.barplot(x='type', y='4770', data=data)
plt.xlabel('type')
plt.ylabel('4770')
plt.title('Tipe Paling laku')
plt.show()
```
